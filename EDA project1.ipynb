## EXPLORATORY DATA ANALYSIS WITH PANDAS, SEABORN AND MATPLOTLIB

In this project i will be carrying out exploratory data analysis on the world population dataset, this dataset has 17 columns;
containing the population of countries in the world from 1970 to 2022 with an interval of 10 years


```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading data
df = pd.read_csv('Desktop/DATASETS/world_population (1).csv')
```

```python
print('shape:', df.shape)
df.head()
```

```python
#changing figures from scientific notation to float
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.sort_values(by = 'World Population Percentage', ascending = False). head(10)
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True), annot = True)

plt.rcParams['figure.figsize'] = (20, 7)

plt.show()
```

```python
df_cont = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only = True).sort_values(by = '2022 Population', ascending = False)
df_cont
```

```python
df_cont_t = df_cont.transpose()
df_cont_t
```

```python
df_cont_t.plot()
```

```python
df_africa = df[df['Continent'] == 'Africa']
df_africa.head()
```

```python
df_africa.sort_values(by = 'World Population Percentage', ascending = False).head(10)
```
